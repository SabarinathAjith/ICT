```python
#Simple ATM Calculator

balance=eval(input("Balance = "))
withdraw=eval(input("Withdrawal = "))
rem=balance-withdraw

if balance>=withdraw:
  print("Transaction successful. Remaining balance:",rem)
else:
  print("Error")
```

```python
#Electricity Bill Generator

units=eval(input("Units consumed = "))
bill=0

if units<=100:
  bill=units * 5
elif units<=200:
  bill=(100*5)+((units-100)*7)
else:
  bill=(100*5)+(100*7)+((units-200)*10)

print("Bill amount = ₹",bill)
```

```python
#Temperature Converter

temp=eval(input("Temp = "))
unit=input("Unit = ")

if unit=="C":
  F=(temp*(9/5))+32
  print("Temperature in Farenheit:",F)
elif unit=="F":
  C=(temp-32)*(5/9)
  print("Temperature in Celsius:",C)
else:
  print("Invalid unit")
```

```python
#Mobile Recharge Validity

from datetime import datetime

format="%Y-%m-%d"
recharge=input("Recharge date: ")
today=input("Today : ")

recharge=datetime.strptime(recharge,format)
today=datetime.strptime(today,format)
diff=today-recharge
passdays=diff.days

if passdays>=28:
    print("Plan expired. Recharge again.")
else:
    remdays=28-passdays
    print(f"Plan is still valid.",remdays,"days remaining.")
```

```python
#Bus Seat Reservation System

booked=eval(input("Booked = "))
cancel=input("Cancel = ")

if cancel in booked:
  booked.remove(cancel)

new=input("New = ")
pos=int(input("Position = "))

booked.insert(pos,new)
print(booked)
```

```python
#Sum of even numbers

n=eval(input("Number: "))
s=0

for i in range(2,n+1,2):
  s=s+i
print("Sum of even numbers:",s)
```

```python
#Dictionary – Student Marks Summary

d={}

for i in range(3):
  name=input("Enter name:")
  marks=float(input("Enter marks:"))
  d[name]=marks

total_marks=sum(d.values())
no_of_students=3
average=total_marks/no_of_students

print("Average marks:",average)
```

```python
#Dictionary – Word Frequency

word_f={}

sentence=input("Sentence:")
sentence_split=sentence.split()

for i in sentence_split:
  if i in word_f:
    word_f[i]+=1
  else:
    word_f[i]=1

for i,count in word_f.items():
  print(i,":",count)

```

```python
#Functions – Check Prime

def prime(p,check_p=True):
  for i in range(2,p):
    if p%i==0:
      check_p=False
      break
  if check_p:
    print(p,"is a prime number")
  else:
    print(p,"is not a prime number")

num=int(input("Enter a number:"))
prime(num)
```

```python
#Vowel Counter

count=0
vowels=["a","e","i","o","u"]
string=input("Enter a string:")
split=list(string)

for i in split:
  if i in vowels:
    count+=1

print("Number of vowels:",count)

```

```python
#Count Divisible Numbers

c=0

N=int(input("Enter the range limit(N):"))
D=int(input("Enter the divisor(D):"))

for i in range(1,N+1):
  if i%D==0:
    c+=1

print(f"Count of numbers divisible by {D} : {c}")
```
